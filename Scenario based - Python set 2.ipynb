```python
# 2.	Scenario: A digital locker requires users to enter a numerical passcode. As part of a security feature, the system checks the sum of the digits of the passcode.
 Write logic to compute the sum of the digits of a given number.

# Read the given number
# convert the number to string then iterate through each character (digit)
# convert it back to an integer, and add it to a running total. 
# sum_digits takes an integer, turns it into a string so it can go through each digit one by one, converts each digit back to an integer, and adds them all up     
```

```python
def sum_digits(number):
    # Convert the number to a string to easily access each digit
    s_number = str(number)
    digit_sum = 0
    # Loop through each character in the string representation of the number
    for digit_char in s_number:
        # Convert the character back to an integer and add it to the sum
        digit_sum += int(digit_char)
    return digit_sum

# Let's test it out!
print(f"Sum of digits for 12345: {sum_digits(12345)}")
print(f"Sum of digits for 901: {sum_digits(901)}")
print(f"Sum of digits for 7: {sum_digits(7)}")
print(f"Sum of digits for 0: {sum_digits(0)}")
```

```python
# In a secure login system, certain features are enabled only for users with prime-numbered user IDs.
 Write logic to check if a given number is prime.

# Read the input number
# number less than 1 is not a prime
# Put the values for i , n is perfectly divisible by i 
# if remainder is there, number is not a prime number
# Enter the number and check whether it is prime or not
```

```python
def is_prime(n):
    factors = 0
    for i in range(1, n+1):
        if n % i == 0:
            factors += 1
    return factors == 2

num =int(input("Enter the number:"))
if is_prime(num):
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# 5.	Scenario: A scientist is working on permutations and needs to calculate the factorial of numbers frequently.
 Write logic to find the factorial of a given number using recursion.

# Read the number
# n is the number for which we want to calculate the factorial.     
# Factorials are traditionally defined only for non-negative integers (0, 1, 2, 3, and so on)
# If n is negative, the function stops immediately and "raises a ValueError." 
# If n is 0 or 1, the function returns 1 and finishes.
# If n is not negative, not 0, and not 1 (meaning n is 2 or a higher positive integer), then this else block is executed
# It calculates the factorial by multiplying n with the factorial of n - 1.
# For the output calls the function     
```

```python
def factorial(n):
    if n < 0:
        raise ValueError  # ("Factorial is not defined for negative numbers")
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
factorial(2)
```

```python
# 6.	Scenario: A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot.
 Write logic to check whether a given number is an Armstrong number.

# Read the number
# Convert the number to a string: This allows easy iteration over each digit.
# Calculate the number of digits
# Compute the sum: To raise each digit to the power of the number of digits and summing them up.
# Compare the sum: Check if the sum equals the original number.
     
```

```python
def is_armstrong_number(number):
    # Convert the number to a string to easily count digits and iterate through them
    num_str = str(number)
    num_digits = len(num_str)
    
    sum_of_powers = 0
    for digit_char in num_str:
        digit = int(digit_char)
        sum_of_powers += digit ** num_digits
        
    # Check if the sum of powers is equal to the original number
    return sum_of_powers == number
```

```python
Scenario: A low-level networking application requires decimal numbers to be converted into binary format before transmission.
 Write logic to convert a given decimal number into its binary equivalent

# Read the number
# An empty string binary_string is created to build the result
# Repeatedly divide the number by 2, storing remainders.
# Reverse the collected remainders to form the binary string.
# Once temp_num becomes 0, the loop terminates, and the binary_string holds the complete binary representation# 
```

```python
def decimal_to_binary(decimal_num):
    # Handle the special case where the input is 0
    if decimal_num == 0:
        return "0"
    
    binary_string = ""
    temp_num = decimal_num
    
    # Keep dividing by 2 until the number becomes 0
    while temp_num > 0:
        # Get the remainder when dividing by 2 (this will be 0 or 1)
        remainder = temp_num % 2
        
        # Prepend the remainder to the binary string
        # We prepend because the remainders are generated in reverse order
        binary_string = str(remainder) + binary_string
        
        # Update the number by integer division by 2
        temp_num = temp_num // 2
        
    return binary_string

# Example usage:
print(decimal_to_binary(100))
print(decimal_to_binary(10)) 
print(decimal_to_binary(25))  
print(decimal_to_binary(12))  
print(decimal_to_binary(150))
```

```python
def decimal_to_binary_short(decimal_num):
    return bin(decimal_num)[2:]

# Example usage:
print(decimal_to_binary_short(128))
print(decimal_to_binary_short(1))
print(decimal_to_binary_short(15))
print(decimal_to_binary_short(10))
print(decimal_to_binary_short(100))
```

```python
# 9.	Scenario: A text-processing tool helps summarize articles by identifying the most significant words.
 Write logic to find the longest word in a sentence.

# Read the sentence
# Split the sentence into words using space as a delimiter
# This might include punctuation attached to words initially
# Clean each word by removing non-alphabetic characters
# And then find the longest among the cleaned words
# Use the max() function with a 'key' to find the item with the maximum length
# Print the longest word
```

```python
def find_longest_word_alt(sentence):
    # Split the sentence into words using space as a delimiter
    # This might include punctuation attached to words initially
    raw_words = sentence.split()

    if not raw_words:
        return "" # Handle empty sentence

    # Clean each word by removing non-alphabetic characters
    # and then find the longest among the cleaned words
    cleaned_words = []
    for word in raw_words:
        # Use a list comprehension or a regex substitute to remove non-alphabetic chars
        # For simplicity, let's just strip common punctuation
        cleaned_word = ''.join(char for char in word if char.isalpha())
        if cleaned_word: # Only add if it's not an empty string after cleaning
            cleaned_words.append(cleaned_word)

    if not cleaned_words:
        return "" # Handle cases where only punctuation existed

    # Use the max() function with a 'key' to find the item with the maximum length
    longest_word = max(cleaned_words, key=len)
    
    return longest_word

# Example usage:
print(find_longest_word_alt("The quick brown fox jumped over the lazy dog."))
print(find_longest_word_alt("A simple sentence."))
print(find_longest_word_alt("No-words-here! 123"))
print(find_longest_word_alt("Hello world"))
print(find_longest_word_alt("shortest and longest"))
print(find_longest_word_alt(""))
print(find_longest_word_alt("!@#$"))
```

```python
import re

def find_longest_word(sentence):
    # Use regular expression to find all words (sequences of letters)
    # This handles punctuation better than just splitting by space
    words = re.findall(r'[a-zA-Z]+', sentence)
    
    if not words:
        return "" # Return empty string if no words are found

    longest_word = ""
    max_length = 0

    for word in words:
        if len(word) > max_length:
            max_length = len(word)
            longest_word = word
            
    return longest_word

# Example usage:
print(find_longest_word("The quick brown fox jumped over the lazy dog."))
print(find_longest_word("A simple sentence."))
print(find_longest_word("No-words-here! 123"))
print(find_longest_word("Hello world"))
print(find_longest_word("shortest and longest"))
print(find_longest_word(""))
print(find_longest_word("!@#$"))
```

```python
# 10.	Scenario: A plagiarism detection tool compares words from different documents and checks if they are anagrams (same characters but different order).
 Write logic to check whether two given strings are anagrams

# Read the two words
# Clean the strings
# Remove non-alphabetic characters and convert to lowercase
# This makes the comparison case-insensitive and ignores punctuation/spaces
# Check lengths
# If the lengths of the cleaned strings are different, they cannot be anagrams
# Sort characters and compare
# Convert strings to lists of characters, sort them, and then compare the sorted lists
# If they are identical, the original strings are anagrams
```

```python
def are_anagrams(str1, str2):
    # Step 1: Clean the strings
    # Remove non-alphabetic characters and convert to lowercase
    # This makes the comparison case-insensitive and ignores punctuation/spaces
    s1_cleaned = ''.join(char for char in str1 if char.isalpha()).lower()
    s2_cleaned = ''.join(char for char in str2 if char.isalpha()).lower()

    # Step 2: Check lengths
    # If the lengths of the cleaned strings are different, they cannot be anagrams
    if len(s1_cleaned) != len(s2_cleaned):
        return False

    # Step 3: Sort characters and compare
    # Convert strings to lists of characters, sort them, and then compare the sorted lists
    # If they are identical, the original strings are anagrams
    return sorted(s1_cleaned) == sorted(s2_cleaned)

# Example usage:
print(are_anagrams("listen", "silent"))         # Expected: True
print(are_anagrams("Debit card", "Bad credit"))   # Expected: True (case and space insensitive)
print(are_anagrams("hello", "world"))           # Expected: False
print(are_anagrams("anagram", "nagaram"))       # Expected: True
print(are_anagrams("A gentleman", "Elegant man")) # Expected: True
print(are_anagrams("", ""))                     # Expected: True (empty strings are anagrams of each other)
print(are_anagrams("a", "a"))                   # Expected: True
print(are_anagrams("a", "b"))                   # Expected: False
print(are_anagrams("" , "abc"))                  # Expected: False
```
