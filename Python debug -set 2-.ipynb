```python
#4.
numbers = [5, 3, 8, 2]
print(numbers.sort())

Output shows 'None', itself does not return the sorted list.
```

```python
numbers = [5, 3, 8, 2]
numbers.sort() # This sorts the list in place, returns None
print(numbers) # This now prints the modified, sorted list
```

```python
# Alternate Method
numbers = [5, 3, 8, 2]
sorted_numbers = sorted(numbers) # This returns a NEW sorted list
print(sorted_numbers)
print(numbers) # The original 'numbers' list is unchanged
```
